# Action Induced Representations (AIR)


This repository contains the **models, losses, and metrics** used in the paper *Disentanglement by Action Induced Representations* [(link)](https://doi.org/10.48550/arXiv.2602.06741)

## Installation

This package is built using [nbdev](https://github.com/AnswerDotAI/nbdev). Clone the repository and install by running the following command:

```bash
pip install /path/to/the/clone
```

### Requirements

The previous command will also ensure that all required libraries are installed. The current version of the code has been developed using the following versions: 

```bash
numpy == 2.2.5
scipy == 1.15.2
fastai == 2.8.1
torch == 2.6.0+cu124
scikit-learn == 1.6.1
jax == 0.6.0
```

## Examples

The `examples/` folder contains Jupyter notebooks that allow you to reproduce the main experimental results of the paper and learn how to train and evaluate VAIR models end-to-end. 

We consider three example datasets:

- **Abstract:** A minimal synthetic environment designed to isolate and study the core inductive biases of AIR models.

- **Classical:** A classical control-style dataset with structured state transitions induced by discrete actions.

- **Quantum:** A dataset based on simulated quantum systems, where actions correspond to interventions on quantum states.
